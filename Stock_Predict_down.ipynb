```python
X = np.array(stock_data.drop(['label', 'rate', 'up_or_down'], axis=1))
X = X[:delta]
y = np.array(stock_data[['up_or_down']])
y = y[:delta].ravel()
for_predict = np.array(future_stock_data.drop(['label', 'rate', 'up_or_down'], axis=1))
y_act = np.array(future_stock_data[['up_or_down']]).ravel()
```

```python
y_pred = mlp.predict(for_predict_scaled)
actual_rate = future_stock_data.rate
actual_rate = actual_rate[:-1].ravel()
try_day = len(y_pred[:-1])
```

```python
#ニューラルネットワークを用いたシュミレーション
invest_money = 0
pocket_money = 1000000
total_return = np.zeros(try_day)
for i in range(0, try_day):
    if y_pred[i] == 1:
        if pocket_money != 0:
            invest_money = pocket_money
            pocket_money = 0
        invest_money = invest_money * actual_rate[i]
    elif y_pred[i] == 0:
        invest_money = invest_money * actual_rate[i]
    elif y_pred[i] == -1:
        if pocket_money == 0:
            pocket_money = invest_money
            invest_money = 0
    total_return[i] = invest_money + pocket_money
print(pocket_money+invest_money)
```

```python
#ずっと保持し続けた場合の株価
non_action = 1000000
for i in range(0, try_day):
    non_action *= actual_rate[i]
non_action
```

```python
#グラフは株価変動
#純利益は初期額100万からの変動額
#投資利益はそのまま株を保持し続けた場合との差額
plt.figure(figsize=(30, 10))
plt.plot(total_return);
print('投資結果:{0}円 → {1:.3f}円'.format(1000000, pocket_money+invest_money))
print('純利益:{:.3f}円'.format(pocket_money+invest_money-1000000))
print('投資利益:{:.3f}円'.format(pocket_money+invest_money-non_action))
print('正答率:{:.3f}'.format(mlp.score(for_predict_scaled, y_act)))
```

```python
up = 0
down = 0
parallel = 0
up_correct = 0
down_correct = 0
para_correct = 0
for i in range(0, try_day):
    if y_act[i] == 1:
        up += 1
        if y_act[i] == y_pred[i]:
            up_correct += 1
    if y_act[i] == -1:
        down += 1
        if y_act[i] == y_pred[i]:
            down_correct += 1
    if y_act[i] == 0:
        parallel += 1
        if y_act[i] == y_pred[i]:
            para_correct += 1

print('Predict Up Count:', up)
print('Predict Up Score:{:.2f}'.format(up_correct/up))
print('Predict Down Count:', down)
print('Predict Down Score:{:.2f}'.format(down_correct/down))
print('Predict Parallel Count:', parallel)
print('Predict Parallel Score:{:.2f}'.format(para_correct/parallel))
```
